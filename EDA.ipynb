Univariate Analysis 

```python

quality_counts = df['quality'].value_counts()
plt.figure(figsize=(8,6))
#print(f' quality count is {quality_counts} and quality index is {quality_counts.index}')

plt.bar(quality_counts.index,quality_counts.values, color='skyblue')
plt.title('Quality Distribution')
plt.xlabel('Quality')
plt.ylabel('Count')
plt.show()

```

Kernel Density Plot: For understanding variance in dataset 

```python
sns.set_style("whitegrid")

numerical_col = df.select_dtypes(include=['float64','int64']).columns
plt.figure(figsize=(12, len(numerical_col) * 2))

for idx, features in enumerate(numerical_col,1):
    plt.subplot(len(numerical_col),2,idx)
    sns.histplot(df[features], kde=True, color='blue')
    plt.title(f'{features} | Skewness: {df[features].skew(): 2f}')

plt.tight_layout()
plt.show()
    
```

```python
# Swarmplot: For understanding distribution of data points across categories
plt.figure(figsize=(10,6))
sns.swarmplot(x='quality', y='alcohol', data=df, palette='viridis')

plt.title('Swarmplot of Alcohol Content by Quality')
plt.xlabel('Quality')
plt.ylabel('Alcohol Content')
plt.show()
```

```python
# BoxPlot: For understanding distribution of data points across categories
plt.figure(figsize=(10,6))
sns.boxplot(x='quality', y='alcohol', data=df, palette='viridis')
```

```python
# Multivariate Analysis 
plt.figure(figsize=(12,8))

sns.heatmap(df.corr(), annot=True, cmap='coolwarm', fmt='.2f', linewidths=2)
plt.title('Correlation Heatmap')
plt.show()
```
